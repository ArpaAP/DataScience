예를 들어, 이전 장에서 우리는 적이 얼마나 많은 전투기를 가지고 있는지 조사했습니다. 선거가 있는 해에는 특정 후보를 지지하는 유권자의 비율이 얼마나 되는지 알고 싶을 수 있습니다. 현재 경제를 평가하기 위해 미국 가구의 연간 중위소득에 관심이 있을 수 있습니다.

이 장에서는 미지의 *모수*를 *추정*하는 방법을 개발할 것입니다. 모수는 모집단과 관련된 수치 값이라는 것을 기억하세요.

모수의 값을 파악하려면 데이터가 필요합니다. 전체 모집단에 대한 관련 데이터가 있다면 단순히 모수를 계산할 수 있습니다.

그러나 모집단이 매우 크다면 – 예를 들어 미국의 모든 가구로 구성되어 있다면 – 전체 모집단에서 데이터를 수집하는 것은 너무 비용이 많이 들고 시간이 오래 걸릴 수 있습니다. 이러한 상황에서 데이터 과학자들은 모집단에서 무작위로 표본을 추출하는 것에 의존합니다.

이것은 추론의 질문으로 이어집니다: 무작위 표본의 데이터를 기반으로 미지의 모수에 대해 정당화할 수 있는 결론을 어떻게 내릴 수 있을까요? 우리는 추론적 사고를 사용하여 이 질문에 답할 것입니다.

무작위 표본을 기반으로 한 통계량은 모집단의 미지의 모수에 대한 합리적인 추정량이 될 수 있습니다. 예를 들어, 표본 가구의 연간 중위소득을 미국 전체 가구의 연간 중위소득에 대한 추정량으로 사용할 수 있습니다.

그러나 모든 통계량의 값은 표본에 따라 달라지며, 표본은 무작위 추출을 기반으로 합니다. 따라서 데이터 과학자들이 무작위 표본을 기반으로 추정량을 제시할 때마다 다음과 같은 질문에 직면하게 됩니다:

**"표본이 다르게 나왔다면 이 추정량은 얼마나 달라질 수 있었을까?"**

이 장에서 여러분은 이 질문에 답하는 한 가지 방법을 배울 것입니다. 이 답은 수치 모수를 추정하고 추정량의 오차량을 정량화하는 도구를 제공할 것입니다.

우리는 백분위수에 대한 예비 지식부터 시작할 것입니다. 가장 유명한 백분위수는 중앙값으로, 소득 데이터 요약에 자주 사용됩니다. 다른 백분위수들은 우리가 개발하려는 추정 방법에서 중요할 것입니다. 따라서 백분위수를 신중하게 정의하는 것부터 시작하겠습니다.
